```python
import matplotlib.pyplot as plt
import pandas as pd
# import torch
```

Ran a local experiment on my laptop - just one epoch. Summary of experiment hyperparameters: 
* batch size 16
* num workers 0
* patch size 256
* Adam optimizer
* lr = 0.0001
* default betas
UNet:
* depth 4
* 32 initial feature maps
* batch norm

The goal was just to see how the learning process resopnds to the given hyperparameters.

```python
train_losses_df = pd.read_csv("../src/output/train_losses.txt",header=None)
fig, ax = plt.subplots()
ax.plot(train_losses_df.index, train_losses_df.iloc[:,0])
ax.set_xlabel("Batches")
ax.set_ylabel("MSE Loss")
ax.set_title("Epoch 1 Training Loss")
```
